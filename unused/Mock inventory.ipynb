```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import kagglehub
```

```python


path = kagglehub.dataset_download("anirudhchauhan/retail-store-inventory-forecasting-dataset")
data = pd.read_csv(path + "/retail_store_inventory.csv")
data.tail()
```

```python
columns_to_keep = [
    "Date",
    "Store ID",
    "Product ID",
    "Category",
    "Region",
    "Inventory Level",
    "Units Sold",
    "Demand Forecast",
    "Holiday/Promotion"
]

# Drop all other columns
cleaned_data = data[columns_to_keep]

# Save cleaned version
cleaned_data.to_csv("inventory_data.csv", index=False)
```

```python
with pd.ExcelWriter("retail_store_inventory_data.xlsx") as writer:
    cleaned_data.to_excel(writer, sheet_name="Inventory", index=False)
```
